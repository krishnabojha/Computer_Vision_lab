# Import required library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Load images

```python
img1 = cv2.imread('./images/copy-1.jpg')
img2 = cv2.imread('./images/copy-2.jpg')
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

# 1.1 Draw keypoints

```python
# 1st Images
sift = cv2.SIFT_create()
kp_1 = sift.detect(img1_gray,None)
img=cv2.drawKeypoints(img1_gray,kp_1,img1)
cv2.imshow('keypoints', img1)
cv2.waitKey(0)
cv2.destroyWindow('keypoints')
```

```python
# 2st Images
sift = cv2.SIFT_create()
kp_2 = sift.detect(img1_gray,None)
img=cv2.drawKeypoints(img1_gray,kp_2,img1)
cv2.imshow('keypoints2', img1)
cv2.waitKey(0)
cv2.destroyWindow('keypoints2')
```

```python
kp_1[0]
```

# 1.2 Features Matching

```python
# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1_gray,None)
kp2, des2 = sift.detectAndCompute(img2_gray,None)

# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)


# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.565*n.distance:
        good.append([m])
# cv.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatchesKnn(img1_gray,kp1,img2_gray,kp2,good,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
len(good)
```

# FLANN-based Matcher

```python
# FLANN parameters
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=50) # or pass empty dictionary
flann = cv2.FlannBasedMatcher(index_params,search_params)
matches = flann.knnMatch(des1,des2,k=2)
# Need to draw only good matches, so create a mask
matchesMask = [[0,0] for i in range(len(matches))]
# ratio test as per Lowe's paper
for i,(m,n) in enumerate(matches):
    if m.distance < 0.565*n.distance:
        matchesMask[i]=[1,0]
draw_params = dict(matchColor = (0,255,0),
 singlePointColor = (255,0,0),
 matchesMask = matchesMask,
 flags = cv2.DrawMatchesFlags_DEFAULT)
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,matches,None,**draw_params)
plt.imshow(img3,),plt.show()
```

# Camera Calibration (Manually)

```python
import numpy as np
from scipy.linalg import svd

# World coordinate points
world_points = np.array([
    [0, -5, 5], [0, -3, 5], [0, -1, 5], [-1, 0, 5],
    [-3, 0, 5], [-5, 0, 5], [0, -5, 3], [0, -3, 3],
    [0, -1, 3], [-1, 0, 3], [-3, 0, 3], [-5, 0, 3],
    [0, -5, 1], [0, -3, 1], [0, -1, 1], [-1, 0, 1],
    [-3, 0, 1], [-5, 0, 1]
])

# Corresponding image coordinate points
image_points = np.array([
    [213.1027, 170.0499], [258.1908, 181.3219], [306.41, 193.8464], [351.498, 183.8268],
    [382.8092, 155.6468], [411.6155, 130.5978], [223.7485, 218.2691], [267.5841, 230.7935],
    [314.5509, 244.5705], [357.7603, 235.1771], [387.819, 205.1184], [415.3728, 178.1908],
    [234.3943, 263.9834], [276.9775, 277.1341], [323.318, 291.5372], [363.3963, 282.1438],
    [392.8288, 251.4589], [419.1301, 223.9051]
])

# Create the homogeneous system of equations
num_points = world_points.shape[0]
A = np.zeros((2 * num_points, 12))

for i in range(num_points):
    X, Y, Z = world_points[i]
    u, v = image_points[i]
    
    A[2 * i] = [X, Y, Z, 1, 0, 0, 0, 0, -u * X, -u * Y, -u * Z, -u]
    A[2 * i + 1] = [0, 0, 0, 0, X, Y, Z, 1, -v * X, -v * Y, -v * Z, -v]

# Perform Singular Value Decomposition
U, D, Vt = svd(A)

# Extract the camera matrix from the last column of V
camera_matrix_manual = Vt[-1].reshape(3, 4)

print("Camera Matrix Obtained Manually:")
print(camera_matrix_manual)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the calibration pattern image
calibration_image = cv2.imread('images/calib_pattern.jpg')

# Image coordinate points
image_points = np.array([
    [213.1027, 170.0499], [258.1908, 181.3219], [306.41, 193.8464], [351.498, 183.8268],
    [382.8092, 155.6468], [411.6155, 130.5978], [223.7485, 218.2691], [267.5841, 230.7935],
    [314.5509, 244.5705], [357.7603, 235.1771], [387.819, 205.1184], [415.3728, 178.1908],
    [234.3943, 263.9834], [276.9775, 277.1341], [323.318, 291.5372], [363.3963, 282.1438],
    [392.8288, 251.4589], [419.1301, 223.9051]
])

# World coordinate points
world_points = np.array([
    [0, -5, 5], [0, -3, 5], [0, -1, 5], [-1, 0, 5],
    [-3, 0, 5], [-5, 0, 5], [0, -5, 3], [0, -3, 3],
    [0, -1, 3], [-1, 0, 3], [-3, 0, 3], [-5, 0, 3],
    [0, -5, 1], [0, -3, 1], [0, -1, 1], [-1, 0, 1],
    [-3, 0, 1], [-5, 0, 1]
])

# Manual Camera Matrix
camera_matrix_manual = np.array([
    [6.48045050e-01,  1.35942226e-01,  5.77325137e-02,  1.43513066e+02],
    [-3.93254390e-02,  7.18778148e-01, -1.68744742e-01,  2.07583456e+02],
    [-2.36397046e-05,  5.67302773e-05,  7.67807200e-04, -1.09027505e-01]
])

# Convert world points to homogeneous coordinates
world_points_homogeneous = np.hstack((world_points, np.ones((world_points.shape[0], 1))))

# Project 3D points to 2D using manual camera matrix
projected_points_manual = np.dot(camera_matrix_manual, world_points_homogeneous.T)
projected_points_manual /= projected_points_manual[2, :]

# Calculate re-projection error for every 2D point
reprojection_errors = []
for i in range(image_points.shape[0]):
    error = np.linalg.norm(projected_points_manual[:2, i] - image_points[i])
    reprojection_errors.append(error)

# Tabulate re-projection error for every 2D point
print("Re-projection Errors for Each 2D Point:")
for i, error in enumerate(reprojection_errors, start=1):
    print(f"Point {i}: {error:.2f} pixels")

# Find the average error of all 2D points
average_error = np.mean(reprojection_errors)
print("\nAverage Re-projection Error:", average_error, "pixels")

# Plot the calibration pattern with red dots for image points and projected points
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(calibration_image, cv2.COLOR_BGR2RGB))
plt.scatter(image_points[:, 0], image_points[:, 1], color='red', label='Provided Points')
plt.scatter(projected_points_manual[0, :], projected_points_manual[1, :], color='blue', marker='x', label='Projected Points (Manual)')
plt.title('Calibration Pattern with Image and Projected Points')
plt.axis('off')
plt.legend()
plt.show()
```

```python
import numpy as np
import cv2
import glob

# Define the size of the checkerboard (number of inner corners)
pattern_size = (6, 9)  # Change this to match your calibration pattern

# Prepare object points (0, 0, 0), (1, 0, 0), ..., (6, 5, 0)
object_points = np.zeros((pattern_size[0] * pattern_size[1], 3), np.float32)
object_points[:, :2] = np.mgrid[0:pattern_size[0], 0:pattern_size[1]].T.reshape(-1, 2)

# Arrays to store object points and image points from all images
obj_points = []  # 3D points in real world space
img_points = []  # 2D points in image plane

# Load calibration images
calibration_images = glob.glob('images/chess.png')  # Change the path to your calibration images

# Detect checkerboard corners in each image and collect points
for image_path in calibration_images:
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    ret, corners = cv2.findChessboardCorners(gray, pattern_size, None)
    
    if ret:
        obj_points.append(object_points)
        img_points.append(corners)

# Get the image size from the first image
image_size = gray.shape[::-1]

# Perform camera calibration using Zhang's method
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(obj_points, img_points, image_size, None, None)

# Test the camera matrix and calculate re-projection error
reprojection_errors = []
for i in range(len(obj_points)):
    img_points_reproj, _ = cv2.projectPoints(obj_points[i], rvecs[i], tvecs[i], mtx, dist)
    error = cv2.norm(img_points[i], img_points_reproj, cv2.NORM_L2) / len(img_points_reproj)
    reprojection_errors.append(error)

# Calculate the average re-projection error
average_error = np.mean(reprojection_errors)

# Print camera matrix and re-projection error
print("Camera Matrix (Intrinsic Parameters):")
print(mtx)
print("\nRe-projection Errors for Each Image:")
for i, error in enumerate(reprojection_errors, start=1):
    print(f"Image {i}: {error:.2f} pixels")

print("\nAverage Re-projection Error:", average_error, "pixels")
```
