## Question 1.2

```python
def compute_homography(correspondence_points):
    A = []
    for src_point, dest_point in correspondence_points:
        x_src, y_src = src_point
        x_dest, y_dest = dest_point
        A.append([-x_src, -y_src, -1, 0, 0, 0, x_dest * x_src, x_dest * y_src, x_dest])
        A.append([0, 0, 0, -x_src, -y_src, -1, y_dest * x_src, y_dest * y_src, y_dest])

    A = np.array(A)

    _, S, Vt = np.linalg.svd(A)
    homography_matrix = Vt[-1].reshape(3, 3)

    return homography_matrix, S, Vt
```

```python
correspondence_points = [
    ([157, 32], [6, 38]),
    ([211, 37], [56, 31]),
    ([222, 107], [82, 87]),
    ([147, 124], [22, 118])
]

homography_matrix, eigenvalues, eigenvectors = compute_homography(correspondence_points)
print("Homography Matrix:")
print(homography_matrix)

print("Eigenvalues:")
print(eigenvalues)

print("Eigenvectors:")
print(eigenvectors)
```

```python
src_point = np.array([x[0] for x in correspondence_points], dtype = np.float32)
des_point = np.array([x[1] for x in correspondence_points], dtype = np.float32)
```

```python
homo_matrix = cv2.getPerspectiveTransform(src_point, des_point, None)
```

## Question 1.6

```python
left_image = cv2.imread('Left_Img.jpg', cv2.IMREAD_GRAYSCALE)
right_image = cv2.imread('Right_Img.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
sift = cv2.SIFT_create()
keypoints_left, descriptors_left = sift.detectAndCompute(left_image, None)
keypoints_right, descriptors_right = sift.detectAndCompute(right_image, None)
```

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(descriptors_left, descriptors_right, k=2)

good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)
```

```python
matched_points_left = np.float32([keypoints_left[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
matched_points_right = np.float32([keypoints_right[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)
```

```python
homography, mask = cv2.findHomography(matched_points_right, matched_points_left, cv2.RANSAC, 5.0)
```

```python
stitched_image = cv2.warpPerspective(right_image, homography, (left_image.shape[1] + right_image.shape[1], left_image.shape[0]))
stitched_image[0:left_image.shape[0], 0:left_image.shape[1]] = left_image
```

```python
cv2.imshow('Stitched Image', stitched_image)
cv2.imshow('Left Image', left_image)
cv2.imshow('Right Image', right_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Question 3

```python
image_paths = ['building1.jpg', 'building2.jpg]

images = [cv2.imread(path) for path in image_paths]

plt.figure(figsize=(10, 1))
for i, image in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"Image {i+1}")
    plt.axis('off')
plt.show()
```

```python
sift = cv2.SIFT_create()

keypoints = []
descriptors = []

for image in images:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    kp, des = sift.detectAndCompute(gray, None)
    keypoints.append(kp)
    descriptors.append(des)
```

```python
matcher = cv2.BFMatcher()
matches = []

for i in range(len(images) - 1):
    matches.append(matcher.knnMatch(descriptors[i], descriptors[i+1], k=2))
```

```python
good_matches = []

for match_pair in matches:
    good = []
    for m, n in match_pair:
        if m.distance < 0.75 * n.distance:
            good.append(m)
    good_matches.append(good)

src_pts = [keypoints[i][m.queryIdx].pt for m in good_matches[i]]
dst_pts = [keypoints[i+1][m.trainIdx].pt for m in good_matches[i]]
```

```python
H, mask = cv2.findHomography(np.float32(src_pts), np.float32(dst_pts), cv2.RANSAC, 5.0)
```

```python
#for panorama dimension
panorama_height = max(image.shape[0] for image in images)
panorama_width = sum(image.shape[1] for image in images)

#for panorama canvas
stitched_panorama = np.zeros((panorama_height, panorama_width, 3), dtype=np.uint8)

#initial  offset
offset = 0


for i in range(len(images) - 1):
    H, _ = cv2.findHomography(np.float32(src_pts), np.float32(dst_pts), cv2.RANSAC, 5.0)
    
    # Warp the current image using the transformation
    warp_img = cv2.warpPerspective(images[i], H, (panorama_width, panorama_height))
    
    # Calculate the width of the portion to place onto the panorama
    paste_width = warp_img.shape[1] - offset
    
    # Place the portion of the warped image onto the stitched panorama
    stitched_panorama[:, offset:offset + paste_width] = warp_img[:, offset:offset + paste_width]
    
    # Update the offset
    offset += images[i].shape[1]
```

```python
plt.figure(figsize=(10, 2))
plt.imshow(cv2.cvtColor(warp_img, cv2.COLOR_BGR2RGB))
plt.title("Stitched Panorama")
plt.axis('off')
plt.show()
```
