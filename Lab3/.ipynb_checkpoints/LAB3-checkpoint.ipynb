## Import required libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Load image and convert it into RGB

```python
image = cv2.imread('./images/wirebond.tif')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## Get structuring element of different sizes

```python
struct_rect11 = cv2.getStructuringElement(cv2.MORPH_RECT, (11,11))
struct_rect15 = cv2.getStructuringElement(cv2.MORPH_RECT, (15,15))
struct_rect45 = cv2.getStructuringElement(cv2.MORPH_RECT, (45,45))

struct_ele15 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (15,15))
struct_ele45 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (45,45))
```

## Erode image with 11x11 size, Rect shape and anchor at (-1,1)

```python
eroded_img11 = cv2.erode(image_rgb, struct_rect11, None,(-1,-1), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img11)
plt.title('Eroded image with 11x11')
```

## Erode with kernel size of 11x11, Rect shape and anchor at (7,10)

```python
eroded_img11 = cv2.erode(image_rgb, struct_rect11, None,(7,10), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img11)
plt.title('Eroded image with 11x11 and centre (7,10)')
```

## Erode with kernel size of 15x15, Rect shape and anchor at (-1,-1)

```python
eroded_img15 = cv2.erode(image_rgb, struct_rect15, None,(-1,-1), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img15)
plt.title('Eroded image with 15x15 and centre (-1,-1)')
```

## Erode with kernel size of 15x15, Rect shape and anchor at (9,2)

```python
eroded_img15 = cv2.erode(image_rgb, struct_rect15, None,(9,2), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img15)
plt.title('Eroded image with 15x15 and centre (9,2)')
```

## Erode with kernel size of 45x45, Rect shape and anchor at (-1,-1)

```python
eroded_img45 = cv2.erode(image_rgb, struct_rect45, None,(-1,-1), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img45)
plt.title('Eroded image with 45x45 and centre (-1,-1)')
```

## Erode with kernel size of 45x45, Rect shape and anchor at (10,5)

```python
eroded_img45 = cv2.erode(image_rgb, struct_rect45, None,(10,5), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img45)
plt.title('Eroded image with 45x45 and centre (10,5)')
```

## Erode with kernel size of 45x45, Ellecptical shape and anchor at (10,3)

```python
eroded_img45_elp = cv2.erode(image_rgb, struct_ele45, None,(10,3), 1, cv2.BORDER_ISOLATED)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(eroded_img45_elp)
plt.title('Eroded image with elleptical shape')
```

## Dilation of Text.tif image

```python
image2 = cv2.imread('./images/text.tif')
image2_rgb2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
# rectangular structuring element
struct_ele_dil1 = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
dilated_image1 = cv2.dilate(image2_rgb2, struct_ele_dil1, None, (-1,-1), 1, cv2.BORDER_ISOLATED)
```

```python
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image2_rgb2)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(dilated_image1)
plt.title('Dilated image')
```

```python
# elliptical structuring element
struct_ele_dil2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
dilated_image2 = cv2.dilate(image2_rgb2, struct_ele_dil2, None, (-1,-1), 1, cv2.BORDER_ISOLATED)
```

```python
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image2_rgb2)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(dilated_image2)
plt.title('Dilated image')
```

```python
#elleptical SE of size 5x5
struct_ele_dil2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
dilated_image2 = cv2.dilate(image2_rgb2, struct_ele_dil2, None, (3,3), 1, cv2.BORDER_ISOLATED)
```

```python
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image2_rgb2)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(dilated_image2)
plt.title('Dilated image')
```

```python
#rectangular SE of size 5x5
struct_ele_dil1 = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
dilated_image1 = cv2.dilate(image2_rgb2, struct_ele_dil1, None, (3,3), 1, cv2.BORDER_ISOLATED)
```

```python
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.imshow(image2_rgb2)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(dilated_image1)
plt.title('Dilated image')
```

## Compute gradient

```python
image_gray = cv2.imread('./images/building.tif', 0)
```

```python
g_x = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
g_y = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

```python
grad_x = cv2.Sobel(image_gray, cv2.CV_32F, 1, 0, None, 3)
grad_y = cv2.Sobel(image_gray, cv2.CV_32F, 0, 1, None, 3)
```

```python
conv_grad_x = cv2.convertScaleAbs(grad_x)
conv_grad_y = cv2.convertScaleAbs(grad_y)
```

```python
final = cv2.add(conv_grad_x, conv_grad_y)
```

```python
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,3,1)
plt.imshow(conv_grad_x)
plt.title("GradX")
fig.add_subplot(1,3,2)
plt.imshow(conv_grad_y)
plt.title('GradY')
fig.add_subplot(1,3,3)
plt.imshow(final)
plt.title('Edge image')
```

## Erosion and boundary extraction

```python
image = cv2.imread('./images/lincoln.tif')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
struct_ele = cv2.getStructuringElement(cv2.MORPH_RECT ,(10,10))
```

```python
eroded_img = cv2.erode(rgb_image, struct_ele, None, (-1,-1), 1, cv2.BORDER_ISOLATED)
```

## Boundary extraction

```python
boundary = rgb_image - eroded_img
```

```python
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.imshow(rgb_image)
plt.title("Original image")
fig.add_subplot(1,3,2)
plt.imshow(eroded_img)
plt.title('Eroded image')
fig.add_subplot(1,3,3)
plt.imshow(boundary)
plt.title('Image Boundary')
```

## Histogram Plotting

```python
image = cv2.imread('./images/fingerprint.tif', 0)
```

```python
histogram = cv2.calcHist([image], [0], None, [256], [0,256])
```

```python
histogram = histogram.ravel()
```

```python
bins = np.arange(256)
plt.figure(figsize=(10, 7))
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.xlim([0, 256])
plt.plot(bins, histogram, color='black')
plt.show()
```

## Binary thresholding

```python
_, thres_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1,2,1)
plt.imshow(image)
plt.title("Original image")
fig.add_subplot(1,2,2)
plt.imshow(thres_image)
plt.title('Binary Thresholding (threshold = 127)')
```
